```python
#1
def is_palindrome(string):
    for i in range(len(string) // 2):
        if string[i] != string[-(i+1)]:
            return False
        return True
```

```python
if is_palindrome(input_str):
    print("PALINDROME")
else:
    print("NOT PALINDROME")
```

```python
#2
def find_the_longest_word(string):
    words = string.split(" ")
    max_len = 0
    for word in words:
        if len(word) > max_len:
            max_len = len(word)
    return max_len
print(find_the_longest_word("dsa dsadaaaaaewqe d dsawewq ddsa   qwe qwdsadasdsadsdsadpewqotowqo eoqwoe qo owqeqow"))
```

```python
#3
import random
def find_num_of_even_and_not(arr):
    even_n = 0
    for item in arr:
        if item % 2 == 0:
            even_n += 1
    return (even_n, len(arr) - even_n)
rand_list = [random.randint(1,1000) for i in range(random.randint(1,1000))]
even, not_even = find_num_of_even_and_not(rand_list)
print(rand_list)
print(even,not_even)
```

```python
#4
some_dict = {'car': 'auto',
             'amazing': 'incredible',
             'anger': 'enrage'}

some_string = "my car is amazing but my neighbour feels anger"
words = some_string.split(" ")
new_string = ""
for word in words:
  new_string += (some_dict[word] if word in some_dict else word) + ' '
print(new_string)
```

```python
#5
def fib(n):
  if n <= 0:
    return 0
  if n == 1:
    return 1
  return fib(n - 1) + fib(n - 2)

fib(10)
```

```python
#6
with open("file.txt", "r") as f:
  lines = f.readlines()
  letters = 0
  len_words = 0
  for line in lines:
    line = line.replace("\n", "")
    words = line.split(" ")
    len_words += len(words)
    for word in words:
      letters += len(word)
  print(f"this text have\n{len(lines)} lines\n{len_words} words\n{letters} letters")
```

```python
#7
def infinity_geometric_progression(b1, q):
    bn = b1
    while True:
        bn = bn * q
        yield bn

geom_gen = infinity_geometric_progression(13, 7)
out = []
for i in range(10):
  out.append(next(geom_gen))
print(out)
```
