```python
import numpy as np
#Отсортировать значения массива по частоте встречания.

np.random.seed(18182)
a = np.random.randint(5, size=10)
print(a)
uniqs, freq = np.unique(a, return_counts=True)
print(uniqs,freq)
indices = np.argsort(freq)
uniqs[indices]
```

```python
#Дана картинка высоты h, ширины w, тип данных np.uint8 (от 0 до 255). Найти количество уникальных цветов.
np.random.seed(18182)
h = 1000
w = 1000
a = np.random.randint(255, size=[3,h,w])


a = np.column_stack([a[0].reshape([1, -1]).transpose(), a[1].reshape([1, -1]).transpose(), a[2].reshape([1, -1]).transpose()])

uniqs, freq = np.unique(a, axis=0, return_counts=True)

len(uniqs)
```

```python
#Написать функцию, вычислающую плавающее среднее вектора

def moving_average(arr):
  return np.multiply(arr, 1 / len(arr))


a = np.array([
    1,2,3,4
])

moving_average(a)
```

```python
#Дана матрица (n, 3). Вывести те тройки чисел, которые являются длинами сторон треугольника.

def is_triangle(arr):
  return (arr[0] + arr[1] > arr[2]) and (arr[1] + arr[2] > arr[0]) and (arr[0] + arr[2] > arr[1])

np.random.seed(18182)
n = 10
a = np.random.randint(255, size=[n,3])
is_tr = np.array(list(map(is_triangle, a)))
print(a)
print(a[np.where(is_tr==True)])
```
