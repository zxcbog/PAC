```python
class Worker:
  '''ABSTRACT CLASS FOR WORKER'''
  def __init__(self, one_hour_pay):
    self.salary = 0
    self.one_hour_pay = one_hour_pay
    self.hours = 0

  def take_salary(self):
    self.salary += 1 * self.hours
    self.hours = 0


class DataScienceWorker(Worker):
  '''DataScientist's worker subclass'''
  def __init__(self, one_hour_pay):
    super().__init__(one_hour_pay=one_hour_pay)
    self.matrix1 = [[]]
    self.matrix2 = [[]]
    self.matrix_out = [[]]

  def get_one_matrix(self, filename):
    with open(filename, "r") as f:
      lines = f.readlines()
      matrix = [[]]
      for i, line in enumerate(lines):
        vals = line.split(" ")
        for val in vals:
          matrix[i].append(int(val))
          if "\n" in val:
            matrix.append([])
    return matrix

  def matrix_get(self, filename1, filename2):
    self.matrix1 = self.get_one_matrix(filename1)
    self.matrix2 = self.get_one_matrix(filename2)

  def matrix_print(self):
    for line in self.matrix_out:
      for val in line:
        print(f"{val} ", end="")
      print()


class Pupa(DataScienceWorker):
  '''PUPA - DEFAULT WORKER(DATA SCIENCE)'''
  def do_work(self, filename1, filename2):
    super().matrix_get(filename1, filename2)
    self.matrix_out = [[0 for j in range(len(self.matrix1[i]))] for i in range(len(self.matrix1))]
    for i in range(len(self.matrix1)):
      for j in range(len(self.matrix1[i])):
        self.matrix_out[i][j] = self.matrix1[i][j] + self.matrix2[i][j]
    self.hours += self.one_hour_pay
    super().matrix_print()


class Lupa(DataScienceWorker):
  '''LUPA - DEFAULT WORKER(DATA SCIENCE)'''
  def do_work(self, filename1, filename2):
    super().matrix_get(filename1, filename2)
    self.matrix_out = [[0 for j in range(len(self.matrix1[i]))] for i in range(len(self.matrix1))]
    for i in range(len(self.matrix1)):
      for j in range(len(self.matrix1[i])):
        self.matrix_out[i][j] = self.matrix1[i][j] - self.matrix2[i][j]
    self.hours += self.one_hour_pay
    super().matrix_print()


class Accountant:
  def give_salary(self, worker: Worker):
    worker.take_salary()

accountant = Accountant()
pupa_obj = Pupa(2)
lupa_obj = Lupa(1)

pupa_obj.do_work("matrix1.txt", "matrix2.txt")
lupa_obj.do_work("matrix1.txt", "matrix2.txt")

accountant.give_salary(pupa_obj)
accountant.give_salary(pupa_obj)
accountant.give_salary(lupa_obj)



print(pupa_obj.salary, lupa_obj.salary)
```

```python
class Queue(list):
  def append(self, val):
    list.append(self, val)

  def pop(self):
    val = list.pop(self, 0)
    return val


class Stack(list):
  def append(self, val):
    list.append(self, val)

  def pop(self):
    return list.pop(self)


class PrivateQue:
  def __init__(self):
    self.__que = Queue()

  def push_back(self, val):
    self.__que.append(val)

  def get_head(self):
    val = self.__que.pop()
    return val

  def get_listque(self):
    return list(self.__que)


que = Queue()
que.append(1)
que.append(2)
que.append(3)
print(que)
stack = Stack()
stack.append(1)
stack.append(2)
stack.append(3)
print(stack.pop())
pque = PrivateQue()
pque.push_back(1)
pque.push_back(2)
pque.push_back(3)
print(pque.get_listque())
print(pque.get_head())
print(pque.get_listque())
```
